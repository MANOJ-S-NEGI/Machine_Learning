taking mean value to fill the Rating nan value

```python
data_copy_rating_mean = round(data_copy['Rating'].mean(),1)
data_copy_rating_mean
```

* data_copy install col need to rectify as it is numeric type showing object
* remove commas and operator sign

```python
data_copy.Installs.head()
```

```python
## replaceing the item
items_need_to_replace = ["+",","]
for i in items_need_to_replace:
    data_copy['Installs'] = data_copy['Installs'].str.replace(i,'');
data_copy['Installs']
```

```python
## checking data type

data_copy['Installs'].dtype
```

```python
## checking null values:

data_copy['Installs'].isnull().sum()
```

```python
## checking wether their is any non numerical values:
data_copy['Installs'].unique()
```

```python
## checking the non numerical:
data_copy[data_copy['Installs']=="Free"]
```

```python
## removing the non numerical value :
data_copy = data_copy.drop(index=10472)
```

```python
data_copy['Installs'].unique()
```

```python
#replaceing the Operend sign:
data_copy['Price'] = data_copy['Price'].str.replace('$','');
```

```python
## converting into numeric:
data_copy['Installs'] = pd.to_numeric(data_copy['Installs'])


data_copy['Reviews'] = pd.to_numeric(data_copy['Reviews'])
data_copy['Price'] = pd.to_numeric(data_copy['Price'])
```

```python
## checking price unique value:
data_copy['Price'].unique()
```

```python
data_copy.info()
```

```python
data_copy.head()
```

```python
##checking unique size:
data_copy.Size.unique()
```

```python
# replacing the M() AND k
data_copy["Size"] = data_copy["Size"].str.replace('M','000')
data_copy["Size"] = data_copy["Size"].str.replace('k','')
```

```python
data_copy.Size.unique()
```

```python
data_copy['Size']=data_copy["Size"].astype("float")
```

```python
data_copy[data_copy["Size"]=='Varies with device']
```

```python
import numpy as np
data_copy["Size"]=data_copy["Size"].str.replace("Varies with device", str(np.nan))
```

```python
## converting data column into float
data_copy['Size']=data_copy["Size"].astype("float")
```

```python
## resizing the data into mb:
for i in data_copy['Size']:
    if i < 10:
        data_copy['Size']=data_copy['Size'].replace(i,i*1000)
data_copy["Size"]=data_copy["Size"]/1000
```

```python
##checking updated columns:
data_copy["Last Updated"]
```

```python
data_copy["Last Updated"]=pd.to_datetime(data_copy["Last Updated"])
```

```python
data_copy.info()
```

```python
data_copy["last_update_day"]=data_copy["Last Updated"].dt.day
data_copy["last_update_month"]=data_copy["Last Updated"].dt.month
data_copy["last_update_year"]=data_copy["Last Updated"].dt.year
```

```python
data_copy.head()
```

```python
data_copy.to_csv("google_cleaned.csv",index=False)
```
